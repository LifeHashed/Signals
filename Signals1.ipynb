Writing the piecewise function

Unit-Step


```python
def step(t):
    return np.piecewise(t, [t < 0, t >= 0], [0, 1])
```

Unit-Pulse

```python
def pulse(t):
  return np.piecewise(t,[t<0,t==0,t>0],[0,1,0])
```

Ramp

```python
def ramp(t):
  return np.piecewise(t,[t<0,t>=0],[0,t])
```

Rectangular Signal


```python
def rectangular_signal(t, duty_cycle=0.5, frequency=1.0, amplitude=1.0):
    T = 1.0 / frequency
    signal = amplitude * (np.mod(t, T) < duty_cycle * T)
    return signal
```

Create Time Values


```python
t = np.linspace(0, 5, 1000)
```

 Generate rectangular signal

```python
duty_cycle = 0.3
frequency = 2.0
amplitude = 1.0
rect_signal = rectangular_signal(t, duty_cycle, frequency, amplitude)
```

Plotting


```python
plt.figure(figsize=(10, 4))
plt.plot(t, rect_signal, label='Rectangular Signal')
plt.title('Rectangular Signal')
plt.xlabel('Time')
```

Asymmetric Sawtooth Waveform

```python
def asymmetric_sawtooth(t, period=1.0, amplitude=1.0, rise_time=0.3, fall_time=0.7):

    t = np.mod(t, period)
    rise_duration = rise_time * period
    fall_duration = fall_time * period

    signal = np.where((t >= 0) & (t < rise_duration), amplitude * t / rise_duration, 0)
    signal += np.where((t >= rise_duration) & (t < period), amplitude * (1 - (t - rise_duration) / fall_duration), 0)

    return signal
```

Generate Time Values

```python
t = np.linspace(0, 2, 1000)
```

Generate Asymmetric Sawtooth Waveform

```python
period = 1.0
amplitude = 1.0
rise_time = 0.3
fall_time = 0.7
asymmetric_sawtooth_waveform = asymmetric_sawtooth(t, period, amplitude, rise_time, fall_time)
```

 Plot the asymmetric sawtooth waveform

```python
 plt.figure(figsize=(10, 4))
plt.plot(t, asymmetric_sawtooth_waveform, label='Asymmetric Sawtooth Waveform')
plt.title('Asymmetric Sawtooth Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

Periodic Gaussian Pulse

```python
def periodic_gaussian_pulse(t, period=1.0, amplitude=1.0, mean=0.0, std_dev=0.1):

    t = np.mod(t, period)
    signal = amplitude * np.exp(-(t - mean)**2 / (2 * std_dev**2))
    return signal
```

Generate Time Values

```python
t = np.linspace(0, 2, 1000)
```

Generate periodic Gaussian pulse

```python
period = 1.0
amplitude = 1.0
mean = 0.5
std_dev = 0.1
periodic_gaussian_pulse_waveform = periodic_gaussian_pulse(t, period, amplitude, mean, std_dev)
```

Plotting

```python
plt.figure(figsize=(10, 4))
plt.plot(t, periodic_gaussian_pulse_waveform, label='Periodic Gaussian Pulse')
plt.title('Periodic Gaussian Pulse')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
