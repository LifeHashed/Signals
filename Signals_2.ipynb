```python
import matplotlib.pyplot as plt
import numpy as np
```

Defining the function

```python
def x(t):
    if 0 <= t < 5:
        return 1
    elif 5 <= t < 8:
        return 2
    elif 8 <= t < 12:
        return 5
    else:
        return 0
```

Generating x(t) plot



```python
# Generate time values
t_values = range(-5,16)  # Adjust the range as needed

# Evaluate x(t) for each time value
x_values = [x(t) for t in t_values]

# Plot the signal
plt.plot(t_values, x_values, label='x(t)')
plt.title('Signal x(t)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

Defining the other function

```python
def y(t):
    if 0 <= t < 7:
        return 2
    elif 7 <= t < 10:
        return 0
    elif 10 <= t < 15:
        return 7
    else:
        return 0
```

Generating y(t) plot

```python
# Generate time values
t_values = range(-5,16)# Adjust the range as needed

# Evaluate y(t) for each time value
y_values = [y(t) for t in t_values]

# Plot the signal
plt.plot(t_values, y_values, label='y(t)')
plt.title('Signal y(t)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

##Performing Operations


1.x(t)+y(t)

```python
# Generate time values
t_values = range(-5,16)  # Adjust the range as needed

# Evaluate x(t) and y(t) for each time value
x_values = [x(t) for t in t_values]
y_values = [y(t) for t in t_values]

# Add x(t) and y(t) to get the sum signal
sum_values = np.add(x_values, y_values)

# Plot the signals
plt.plot(t_values, x_values, label='x(t)')
plt.plot(t_values, y_values, label='y(t)')
plt.plot(t_values, sum_values, label='x(t) + y(t)', linestyle='--', linewidth=2)
plt.title('Sum of x(t) and y(t)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

2.x(t)-y(t)

3.x(t)*y(t)

```python
# Generate time values
t_values = range(-5,16)  # Adjust the range as needed

# Evaluate x(t) and y(t) for each time value
x_values = [x(t) for t in t_values]
y_values = [y(t) for t in t_values]

# Subtract y(t) from x(t) to get the difference signal
diff_values = np.subtract(x_values, y_values)

# Multiply x(t) and y(t) to get the product signal
prod_values = np.multiply(x_values, y_values)

# Plot the signals
plt.plot(t_values, x_values, label='x(t)')
plt.plot(t_values, y_values, label='y(t)')
plt.plot(t_values, diff_values, label='x(t) - y(t)', linestyle='--', linewidth=2)
plt.plot(t_values, prod_values, label='x(t) * y(t)', linestyle='-.', linewidth=2)
plt.title('Subtraction and Multiplication of x(t) and y(t)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

4. x(t)/2+y(t)/3

```python
# Generate time values
t_values = range(-5,16)  # Adjust the range as needed

# Evaluate x(t) and y(t) for each time value
x_values = np.array([x(t) for t in t_values])
y_values = np.array([y(t) for t in t_values])

# Compute the expression (x(t)/2) + (y(t)/3)
result_values = (x_values / 2) + (y_values / 3)

# Plot the signals
plt.plot(t_values, x_values, label='x(t)')
plt.plot(t_values, y_values, label='y(t)')
plt.plot(t_values, result_values, label='(x(t)/2) + (y(t)/3)', linestyle='-.', linewidth=2)
plt.title('Expression (x(t)/2) + (y(t)/3)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
