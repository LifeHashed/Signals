## Write a program to convolve two discrete-time sequences. Plot all the sequences. Verify the result by analytical calculation.


Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining the function

```python
def convolution(x, h):
    # Convolve sequences x and h
    return np.convolve(x, h, mode='full')
```

Plot function

```python
def plot_sequence(title, sequence, xlabel, ylabel):
    plt.stem(sequence, use_line_collection=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

 Define two discrete time sequences

```python
# Define two discrete time sequences
t=np.arange(-10,10,1)
x = np.sin(t)  # Sequence 1
h = np.cos(t)  # Sequence 2
```

 Convolve the sequences

```python

y_convolved = convolution(x, h)
```

Plot of the original sequence

```python

plot_sequence('Sequence x', x, 'n', 'x[n]')
plot_sequence('Sequence h', h, 'n', 'h[n]')
```

Plot the convolution result

```python

plot_sequence('Convolution Result (Analytical)', y_convolved, 'n', 'y[n]')
```

Verify the result using analytical calculations

```python

y_analytical = np.convolve(x, h, mode='full')
assert np.array_equal(y_convolved, y_analytical), "Verification failed"

print("Verification successful!")

# Plot the convolution result using numpy's convolve for comparison
y_numpy = np.convolve(x, h, mode='full')
plot_sequence('Convolution Result (Numpy)', y_numpy, 'n', 'y[n]')
```

## Find the convolution of two Non Causal Signal x(n) = 3δ(n + 2) − δ(n − 1) + 2δ(n −3) and h(n) = u(n + 4) − u(n − 3). Plot all the sequences. The value of n should be 50.

Define the signals



```python

def x_n(n):
    return 3 * (n == -2) - (n == 1) + 2 * (n == 3)

def h_n(n):
    return (n >= -4).astype(int) - (n >= 3).astype(int)
```

Computing


```python
# Define the range of values for n
n_values = np.arange(-50, 51)

# Compute the convolution y(n) = x(n) * h(n)
y_n = np.convolve(x_n(n_values), h_n(n_values), mode='same')
```

Plotting


```python
plt.stem(n_values, x_n(n_values), label='x(n)', use_line_collection=True)
plt.stem(n_values, h_n(n_values), label='h(n)', use_line_collection=True)
plt.stem(n_values, y_n, label='y(n)', use_line_collection=True)
plt.title('Signal Sequences and Convolution')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

## Generate a one-dimensional signal , which may be a sinusoidal 𝑥 signal or any random signal with − 2 < 𝑥 < 2. Add an additive random noise and find the response of an LTI system on 𝑥 whose impulse response ℎ1 = 𝑜𝑛𝑒𝑠(1,3)/3 and ℎ2 = 𝑜𝑛𝑒𝑠(1,7)/7 ℎ3= 𝑜𝑛𝑒𝑠(1,11)/11

#also analyze:

```python


# Generate a one-dimensional signal x(n) (sinusoidal signal)
n_values = np.arange(0, 100)
x_n = np.sin(2 * np.pi * 0.05 * n_values)  # You can replace this with any signal
```

```python


# Add random noise to x(n)
noise = np.random.uniform(-2, 2, size=len(n_values))
x_n_with_noise = x_n + noise
```

```python


# Define impulse responses h1(n), h2(n), and h3(n)
h1_n = np.sin(2 * np.pi * 0.05 * 3 * n_values) / 3
h2_n = np.sin(2 * np.pi * 0.05 * 7 * n_values) / 7
h3_n = np.sin(2 * np.pi * 0.05 * 11 * n_values) / 11
```

```python


# Convolve x(n) with each impulse response
y1_n = np.convolve(x_n_with_noise, h1_n, mode='same')
y2_n = np.convolve(x_n_with_noise, h2_n, mode='same')
y3_n = np.convolve(x_n_with_noise, h3_n, mode='same')
```

```python


# Plot x(n) and y(n) for each impulse response
plt.plot(n_values, x_n_with_noise, label='x(n) with noise')
plt.plot(n_values, y1_n, label='y1(n) (h1 convolution)', linestyle='dashed')
plt.plot(n_values, y2_n, label='y2(n) (h2 convolution)', linestyle='dashed')
plt.plot(n_values, y3_n, label='y3(n) (h3 convolution)', linestyle='dashed')
plt.title('Signal and Convolution Responses')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
